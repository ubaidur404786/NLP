```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./dataset/100_Unique_QA_Dataset.csv')

df.head()
```

```python
def tokenize(text):
    text= text.lower()
    text = text.replace('?','')
    text = text.replace("'","")
    return text.split()
```

```python
tokenize('What is the capital of France?')
```

```python
#vocab
vocab={'<UNK>':0}
def build_vocab(row):
    
    tokenized_q = tokenize(row['question'])
    tokenized_a = tokenize(row['answer'])
    merged_tokebnized = tokenized_q + tokenized_a
    

    for token in merged_tokebnized:
        if token not in vocab:
            vocab[token]=len(vocab)
    
     
```

```python
df.apply(build_vocab,axis=1)

```

```python
vocab
```

```python
#covert text to numerical indices
def text_to_indices(text,vocab):

    indexed_text=[]
    for token in tokenize(text):
        if token in vocab:
            indexed_text.append(vocab[token])
        else:
            indexed_text.append(vocab['<UNK>'])  # Append index for unknown token
    return indexed_text
```

```python
text_to_indices('Capital of Franceeee ?',vocab)
```
